# showing F = ma with pulleys and lightgates

By setting up an experement as seen in diagram below, we can get data point of force (mg) against acceleration
Then we can plot the force agains the mass to show that $F \propto a$ with constant mass, and further calculate the mass as $m = \frac{F}{a}$

---
## meth
---

1. Fix the tread to the rider and attach 8 slotted 50 gram masses
2. Set a pair of light gates 50 cm away to record the speed as it passes through
3. Use plates under the end of the rail opposite to the pully to prduce a friction compencating slope
4. record the masses used on the pully, and the speed at which it passed through the lightgates
5. remove a mass, and add to the rider, such that total mass M+m is constant, for easier calculations
6. repeat the experement untill but 1 mass is on the pully
7. plot force against acceleration to find a graph which should show the propotionality of force against acceleration





```python
import scipy.stats as stats
import matplotlib.pyplot as plt
import numpy as np
g = 9.81 #m/s^2
u = 0
masses = [0,0.05, 0.1, 0.15, 0.2, 0.25, 0.3, 0.35, 0.4] #kg
v = [0,0.89, 1.24, 1.48 , 1.8, 1.95, 2.14, 2.26, 2.4] #m/s
s = 0.5 #m
#v^2 = u^2 + 2as
a = []
for i in range(len(v)):
    a.append((v[i]**2 - u**2)/(2*s))
forces = []
for i in range(len(masses)):
    forces.append(masses[i]*9.81)

slope, intercept, r, p, std_err = stats.linregress(a, forces)
plt.scatter(a, forces)
plt.ylabel("Force (N)")
plt.xlabel("Acceleration (m/s^2)")
plt.plot(a, slope*np.array(a) + intercept)


print("mass =", slope, "kg")
```
