# Determination of Young modulus of a metal 
### Theory:
Youngs modulus = $\frac{stress}{strain}$ or $E = \frac{F/A}{\Delta l / l} = \frac{F l}{\Delta l A} $

if we plot a graph of stress against strain then
$ E = m\frac{l}{A} $

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

d = 0.245*10**(-3)
a = (d/2)**2*np.pi
masses = np.array(range(1,8,1))*0.1
extentionsmm = np.array([445, 445, 447, 449, 450, 451, 453])
extentions = extentionsmm*(10**(-3))
extensions = extentions - extentions[0]
stress = masses*9.81/a
strain = extensions/extentions[0]
#slope, intercept, r_value, p_value, std_err = stats.linregress()
plt.scatter(strain, stress)
plt.xlabel("strain")
plt.ylabel("stress")
slope, intercept, r_value, p_value, std_err = stats.linregress(strain, stress)
plt.plot(strain, slope*strain+intercept)
print(slope)


```
