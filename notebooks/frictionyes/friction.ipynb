# Experiment: Measuring Carpet Friction by Dropping Cars Down a Ramp

## Introduction
In this experiment, we investigated the friction of a carpet by dropping cars down a ramp and measuring how far they traveled along the carpet from different heights up the ramp. The distance traveled by the cars can give us insights into the frictional forces acting on the cars as they move along the carpet.

## Materials
- Cars
- Ramp
- Carpet
- Measuring tape

## Procedure
1. Set up the ramp at a suitable angle.
2. Place the carpet at the end of the ramp to provide a surface for the cars to travel on.
3. Measure the height of the ramp from which the cars will be dropped.
4. Drop a car from the measured height and start the stopwatch.
5. Observe and measure the distance traveled by the car along the carpet using the measuring tape.
7. Repeat steps 4-6 for different heights of the ramp.

## Data Analysis
Once the data is collected, we can analyze it to determine the relationship between the height of the ramp and the distance traveled by the cars along the carpet. This relationship can help us understand the frictional forces acting on the cars.

## Conclusion
By conducting this experiment, we can gain insights into the friction of the carpet and its impact on the motion of the cars. The data collected can be used to further study and analyze the factors affecting friction and develop strategies to reduce friction in various applications.


```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

ramplengths = np.array([0.8, 0.75, 0.7, 0.65, 0.6, 0.55, 0.5, 0.45, 0.4, 0.35, 0.3, 0.25, 0.2, 0.15])
distances = np.array([2.12, 2.1, 1.76, 1.55, 1.35, 1.3, 1.17, 1.27, 0.98, 0.8, 0.66, 0.49, 0.24, 0.16])
heights = ramplengths*48/84
u = (9.81*heights*2)**0.5

slope, intercept, r_value, p_value, std_err = stats.linregress(u, distances)
plt.scatter(u, distances)
plt.plot(u, slope*u+intercept)
plt.show()
```
