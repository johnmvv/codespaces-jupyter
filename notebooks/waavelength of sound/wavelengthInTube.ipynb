# determining the speed of sound using stationary waves

By setting up a speaker, connected to a frequency generator, held above a measuring cylinder, by varying the height of the water, a node of the stationary wave can be found at the height at which the sound is loudest. 

The wavelength can be measued as being double the distance between the nodes, as this is the half wavelength.
as $c = f\lambda$ 

if d is the distance between nodes then $c = 2fd$ 

so  $2d = c/f$

as such we can plot a graph of $2d$ against $\frac{1}{f}$, to find a line with gradient $c$


```python
import scipy.stats as stats
import matplotlib.pyplot as plt
import numpy as np


frequency = np.array([1200, 1500, 2500])
dx2 = np.array([0.282, 0.237, 0.124])

x = np.array([1/i for i in frequency])
y = dx2

for i in range(0, 3):
    x = np.append(x, 0)
    y = np.append(y, 0)


slope, intercept, r, p, std_err = stats.linregress(x,y)
plt.scatter(x, y)
plt.plot(x, slope*np.array(x) + intercept)

print("speed =", slope, "m/s")
```
