# Calculating the weight of a ruler using moments


```python
from matplotlib import pyplot as plt
from scipy import stats

x = [0.0575,0.11,0.165,0.220]
y = [0.00425, 0.0090, 0.0135,0.0180]
#calling the linregress function for info stuff
slope, intercept, r, p, std_err = stats.linregress(x, y)
#loop to calculate the line of best fit
line=[]
for i in range (len(x)):
    line.append(slope*x[i]+intercept)


plt.scatter(x,y)
plt.plot(x,line)
plt.ylabel('moment (nm)')
plt.xlabel('distance (m)')
plt.axis([0,0.220,0,0.0180])
plt.title('Moment vs. Distance')
plt.grid(True)
plt.show()
```

```python
#finding mass of the ruler
weight = slope
print(weight,'N')
mass = weight/9.81
print(mass,'kg')
```
