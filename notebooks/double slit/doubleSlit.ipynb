# double slit wavelength
## theory
The space between the fringes $\Delta y$ is given by the equation $\frac{wavelength*distance}{apature}$ or $\frac{\gamma D}{d}$
As wavelength is constant, and apature remains unchanged, a graph of axis D and $\Delta y$ can be drawn, with a gradient of $\frac{\gamma}{d}$

## Aparaturs
* Laser pen
* stand
* clamp
* double slit medium
* meter rule
* ruler

## Method
The aparatus should be set up as in the diagram, with the distance varied for new measurements

![](image_2023-10-05_094656331.png)

the distance D should be measured with a meter rule, and the spacing $\Delta y$ should be measured with a ruler or higher resolution fromthe centeral fringe to the next closest fringe, or in counts of 10, for higher accuracy.

```python
#double slit experement
import numpy as np
import matplotlib.pyplot as plt
distances = np.array([1, 2, 3, 4]) #m
measured = np.array([0.001, 0.001, 0.002, 0.003]) #m+/-0.001m
averagey = np.average(measured)
averaged = np.average(distances)
a = 0.001 #m
#wavelength = ay/d
waveLength = a*averagey/averaged #m
waveLengthinnm = waveLength*1000000000 #nm
print('WaveLength =', waveLengthinnm, 'nm')
plt.plot(distances, measured, 'o')
plt.plot([0,1*4],[0,averagey/averaged*4], '-')
plt.xlabel('Distance (m)')
plt.ylabel('Measured (m)')
plt.title('Double Slit Experiment')
plt.show()
```
