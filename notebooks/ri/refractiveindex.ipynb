```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

Ai = [0,15,30,45,60,75]
Ar = [0, 9, 20, 26, 36, 40]


sini = np.sin(np.deg2rad(Ai))
sinr = np.sin(np.deg2rad(Ar))

slope, intercept, r_value, p_value, std_err = stats.linregress(sinr, sini)

plt.plot(sinr, sini, 'o', label='original data')
plt.plot(sinr, intercept + slope*sinr, 'r', label='fitted line')
plt.ylabel('sin(i)')
plt.xlabel('sin(r)')
plt.show()

print("slope: ", slope)

```

# RI Physics Practical

## Introduction
In this physics practical, we will be investigating the phenomenon of refraction and the concept of refractive index (RI). Refraction occurs when light passes from one medium to another and changes direction due to the change in speed. The refractive index is a measure of how much the speed of light is reduced in a particular medium compared to its speed in a vacuum.

## Aim
The aim of this practical is to determine the refractive index of a given material.

## Materials
- Laser pointer
- Semicircular glass block
- Protractor
- Ruler
- Paper

## Procedure
1. Set up the experimental apparatus by placing the semicircular glass block on a flat surface.
2. Place a sheet of paper underneath the glass block to make it easier to observe the laser beam.
3. Turn on the laser pointer and direct the beam towards the curved surface of the glass block.
4. Use the protractor to measure the angle of incidence (i) between the laser beam and the normal to the surface of the glass block.
5. Observe the laser beam as it enters the glass block and measure the angle of refraction (r) between the laser beam and the normal inside the glass block.
6. Repeat steps 3-5 for different angles of incidence.
7. Record the values of i and r for each measurement.

## Data Analysis
1. Calculate the sine of the angle of incidence (sin(i)) and the sine of the angle of refraction (sin(r)) using the measured values.
2. Plot a graph of sin(i) against sin(r).
3. Use the slope of the graph to calculate the refractive index (RI) of the glass block using the formula: RI = sin(i) / sin(r).
4. Calculate the average value of the refractive index from multiple measurements.

## Conclusion
In this practical, we successfully determined the refractive index of the given glass block using the phenomenon of refraction. The calculated value of the refractive index provides valuable information about the optical properties of the material. This practical demonstrates the relationship between the angles of incidence and refraction and how it relates to the refractive index of a medium.
